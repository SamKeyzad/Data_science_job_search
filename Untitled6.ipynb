```python
import pandas as pd
import numpy as np
```

```python
job_postings = pd.read_csv(r'/Users/hassankazemzade/Downloads/archive-2/job_postings.csv')
job_skills = pd.read_csv(r'/Users/hassankazemzade/Downloads/archive-2/job_skills.csv')
```

```python
print(job_postings.head())
print(job_postings.info())
```

```python
print(job_skills.head())
print(job_skills.info())
```

```python
job_postings.dropna(inplace=True)
job_skills.dropna(inplace=True)
```

```python
merged_df = pd.merge(job_postings ,job_skills , on='job_link', how = 'inner' )
```

```python
merged_df
```

```python
merged_df['job_skills'] = merged_df['job_skills'].str.lower()
```

```python
import re
```

```python
merged_df['job_skills'] = merged_df['job_skills'].apply(lambda x: re.sub(r'[^\w\s,]', '', str(x)))
```

```python
merged_df['job_skills'] = merged_df['job_skills'].str.split(',')
```

```python
merged_df['job_skills']
```

```python
all_skills = merged_df['job_skills'].explode().str.strip()
```

```python
all_skills
```

```python
from collections import Counter
skill_counts = Counter(all_skills)
```

```python
skill_counts
```

```python
skill_counts_df = pd.DataFrame(skill_counts.items(), columns=['Skill', 'Count'])
```

```python
skill_counts_df
```

```python
top_skills = skill_counts_df.sort_values(by='Count', ascending=False)
```

```python
top_skills
```

```python
print(top_skills.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_n = 9
top_skills = skill_counts_df.sort_values(by='Count', ascending=False).head(top_n)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Count' , y = 'Skill' , data = top_skills , palette ='viridis' )
plt.title(f'top {top_n} most common data scientists skills ' , fontsize= 16)
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Skills',fontsize=12)
plt.show
```
